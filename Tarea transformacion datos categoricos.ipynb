```python
import pandas as pd
```

```python
adidas = pd.read_csv("adidas.csv")
```

```python
adidas.columns
```

```python
datos_categoricos = ['url', 'name', 'sku', 'currency',
       'availability', 'color', 'category', 'source', 'source_website',
       'breadcrumbs', 'description', 'brand', 'images', 'country', 'language',
       ]
```

Eliminar los datos que tengan mas del .98%


```python
target = .97

columnas_validas = []
columnas_arriba_98 = []
columnas_un_nivel = []

for c_ in datos_categoricos:
    x = adidas.loc[ : , [c_] ].drop_duplicates()
    uniquevalues = x.shape[0]
    
    if uniquevalues == 1:
        columnas_un_nivel.append(c_)
    elif (uniquevalues / adidas.shape[0]) > target:
        columnas_arriba_98.append(c_)
    else:
        columnas_validas.append(c_)
        
    
    
```

```python
adidas.loc[ : , columnas_arriba_98 ].drop_duplicates()
```

```python
adidas.loc[ : , columnas_validas ].drop_duplicates()
```

Tratamiento de la columna breadcrumbs

```python
columnas_validas
```

```python
frecuencia = adidas.groupby( [ "breadcrumbs" ] , as_index = False ).agg({"sku" : [ "count" ] })
```

```python
frecuencia.columns = frecuencia.columns.droplevel(1)
```

```python
frecuencia['Porcentage'] = frecuencia['sku'] / adidas.shape[0]
```

```python
frecuencia = frecuencia.sort_values( ['Porcentage'] , ascending=False )
```

```python
frecuencia #Ya que las primeras 4 filas forman el 71% del dataset podemos tomar a todo lo demas que esta por debajo de ella como Others 
```

```python
frecuencia['breadcrumbs_p'] = 'Others'
```

```python
frecuencia
```

```python
frecuencia['breadcrumbs_p'] = frecuencia.apply( lambda r: r['breadcrumbs'] if r['Porcentage'] > 0.10 else r['breadcrumbs_p'] ,axis = 1 )
```

```python
frecuencia
```

```python
frecuencia = frecuencia.loc[ : , ['breadcrumbs','breadcrumbs_p'] ]
```

```python
adidas.merge( frecuencia , on=['breadcrumbs'] , how='left' )
```
